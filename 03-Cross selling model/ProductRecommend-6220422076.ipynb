```python
import pandas as pd

df = pd.read_csv('/content/sample_data/Customer Preference Survey (Responses) - Form Responses 1.csv')
```

```python
df.head()
```

```python
df_drop_na = df.dropna()
```

```python
df_drop_na.head()
```

```python
df_drop_na.drop(['Timestamp'], axis=1, inplace=True)
```

```python
df_drop_na.head()
```

```python
df_drop_na.dropna(axis=0,inplace=True)
```

```python
df_drop_na
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets = apriori(df_drop_na, min_support=0.1, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(10)
```

Rule ที่พบจาก data set ชุดนนี้คือคนที่ซื้อเครื่องซักผ้า มันจะซื้อเครื่องอบผ้าคู่กันด้วยครับ
